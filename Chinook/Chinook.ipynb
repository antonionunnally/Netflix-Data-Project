**1. How many customers patronize the media store?**

```python
conn = sqlite3.connect("/content/Chinook.sqlite")
cursor = conn.cursor()
cursor = conn.execute("SELECT COUNT(DISTINCT customerid) as 'number of customers' FROM customer")
rows = cursor.fetchall()
rows
```

**2. How many orders have been made?**

```python
conn = sqlite3.connect("/content/Chinook.sqlite")
cursor = conn.cursor()
cursor = conn.execute("SELECT COUNT(DISTINCT invoiceid) as 'number of orders'  FROM invoice")
rows = cursor.fetchall()
rows
```
